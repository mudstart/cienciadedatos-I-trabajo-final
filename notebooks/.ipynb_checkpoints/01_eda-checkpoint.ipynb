## 0. Conexión a la base de datos y diagnóstico

```python
import sys, os
sys.path.append('../src')
sys.path.append('..')
import pandas as pd, numpy as np
from sqlalchemy import inspect
from db import crear_engine
pd.set_option('display.max_columns', 60)

engine = crear_engine(silencioso=False)
assert engine is not None, 'No hay conexión a MySQL. Revisa src/config.py.'
print('Conexión establecida.')
```

```python
insp = inspect(engine)
tablas = insp.get_table_names()
resumen = [{'tabla': t,
            'filas': int(pd.read_sql(f'SELECT COUNT(*) n FROM `{t}`', engine)['n'].iloc[0]),
            'columnas': len(insp.get_columns(t))} for t in tablas]
pd.DataFrame(resumen)
```

## 0b. Validación del esquema OULAD

```python
import validar_bd
problemas = validar_bd.validar(engine)
assert problemas == 0, 'El esquema OULAD tiene problemas; revisa el detalle arriba.'
```

## 1. Obtain + Scrub

```python
from pipeline import PipelineOULAD
pipe = PipelineOULAD(engine=engine, verbose=True).obtain().scrub()
pipe.df.head()
```

```python
print('Distribución de final_result:')
print(pipe.df['final_result'].value_counts())
print('\nMissing imputados:', pipe.reporte_eda.get('na_antes', {}))
print('Tabla analítica:', pipe.df.shape)
```

## 2. Explore — descriptiva, kurtosis, asimetría, correlación, boxplot, scatter

```python
pipe.explore()
pipe.reporte_eda['describe'].round(3)
```

```python
kurt = pipe.reporte_eda['kurtosis'].rename('kurtosis')
skew = pipe.reporte_eda['skew'].rename('skew')
pd.concat([kurt, skew], axis=1).round(3)
```

```python
from IPython.display import Image, display
for f in ['hist_score.png','box_vle_result.png','correlacion.png','scatter_vle_score.png']:
    p = os.path.join('../outputs/figures', f)
    if os.path.exists(p): display(Image(p))
```

## 3. Model — 3 algoritmos supervisados por target + KMeans

```python
pipe.model()
# Vista rápida de métricas de los modelos OULAD
filas = []
for tarea in ['clasificacion_binario','clasificacion_ordinal','regresion_score']:
    for alg, m in pipe.resultados[tarea]['metricas'].items():
        for k, v in m.items():
            if k != 'f1_manual': filas.append({'tarea':tarea,'modelo':alg,'metrica':k,'valor':v})
pd.DataFrame(filas).round(4)
```

## 4. PI2 — experimento complementario Kongo (binario) con msePI2 / r2PI2

```python
pipe.pi2()
p2 = pipe.resultados['pi2_kongo']; mejor2 = p2['mejor']
print('Mejor modelo PI2:', mejor2)
pd.Series(p2['metricas'][mejor2])
```

## 5. iNterpret — reporte consolidado, f1 manual e importancias

```python
pipe.interpret()   # guarda outputs/reporte_metricas_general.csv
pipe.reporte_metricas.round(4)
```

```python
info = pipe.resultados['clasificacion_binario']
print('Mejor clasificador dicotómico:', info['mejor'])
pd.Series(info['metricas'][info['mejor']]['f1_manual'])
```

```python
imp = pd.Series(info['importancias']).sort_values(ascending=False)
imp.head(15).round(4)
```

```python
for f in ['cm_binario_%s.png' % info['mejor'],'importancias_binario.png','importancias_regresion.png']:
    p = os.path.join('../outputs/figures', f)
    if os.path.exists(p): display(Image(p))
```

## 6. Guardar resumen estructurado

```python
# Guarda EDA + métricas (msePI2/r2PI2 incluidos) en un solo archivo
pipe.exportar_resumen()
print('Listo: outputs/resumen_resultados.json')
```
